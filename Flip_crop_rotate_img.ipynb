IMAGE ROTATE, CROP, AND FLIP 
 

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./kitty.jpg")
#print(image)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# plt.imshow(image)
plt.imshow(image_rgb)
```

Flipping

```python
#1 -- flip hrizontaly
#0 -- flip vertically
# -1 -- flip both axis
flip_horizontal=cv2.flip(image, 1)
flip_vertical= cv2.flip(image, 0)
flip_both_axis = cv2.flip(image, -1)


titles = ["original", "horizontal", "vertical", "both axis"]
images = [image_rgb, 
          cv2.cvtColor(flip_horizontal, cv2.COLOR_BGR2RGB),
          cv2.cvtColor(flip_vertical, cv2.COLOR_BGR2RGB),
          cv2.cvtColor(flip_both_axis, cv2.COLOR_BGR2RGB) 
          ]

plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

IMAGE ROTATE 


```python
rotate_90_clock= cv2.rotate(image_rgb, cv2.ROTATE_90_CLOCKWISE)
rotate_180=  cv2.rotate(image_rgb, cv2.ROTATE_180)
rotate_90_counterclock= cv2.rotate(image_rgb, cv2.ROTATE_90_COUNTERCLOCKWISE)

titles = ["original", "90_clock", "_180_", "90_counterclock"]
images = [image_rgb, rotate_90_clock, rotate_180, rotate_90_counterclock]

plt.figure(figsize=(12, 8))

for i in range(4):
    plt.subplot(2,2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()

```
